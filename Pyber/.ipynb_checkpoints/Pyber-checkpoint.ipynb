### Takeaways
* Drivers in Rural cities make significantly more per ride than those in Urban cities. Distance of trip was not factored into this analysis, so that may be a cause. Another potential cause is the significantly larger population of drivers in Urban cities compared to Rural

* Rides in rural cities have a significantly larger variance in average fare compared to those in Suburban or Urban cities

* For drivers, Suburban cities seem to be the best bang for buck. They have similar volume of rides to a number of Urban cities while consistently having a higher average fare

```python
%matplotlib inline
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

citydata = pd.read_csv("data/city_data.csv")
ridedata = pd.read_csv("data/ride_data.csv")

print(citydata.head())
print(ridedata.head())
```

```python
merged = pd.merge(citydata, ridedata, on="city")
merged.head()
```

```python
merged.dtypes
```

```python
merged.head()
```

```python
# Obtain the x and y coordinates for each of the three city types

# Extract City Type
urban = merged.loc[merged["type"] == "Urban"]
suburban = merged.loc[merged["type"] == "Suburban"]
rural = merged.loc[merged["type"] == "Rural"]

# Group by city to find key metrics per city
u_grouped = urban.groupby("city")
s_grouped = suburban.groupby("city")
r_grouped = rural.groupby("city")

# Find total number of rides per city and pass into list
u_rides = u_grouped["ride_id"].count()
u_ridelist = [x for x in u_rides]
s_rides = s_grouped["ride_id"].count()
s_ridelist = [x for x in s_rides]
r_rides = r_grouped["ride_id"].count()
r_ridelist = [x for x in r_rides]

# Find average fare per city and pass into list
u_fares = u_grouped["fare"].mean()
u_farelist = [x for x in u_fares]
s_fares = s_grouped["fare"].mean()
s_farelist = [x for x in s_fares]
r_fares = r_grouped["fare"].mean()
r_farelist = [x for x in r_fares]

# Find driver count per city and pass into list
# Multiplied each cities drivercount by 10 to highlight the relative difference in driver count per city
u_drivercounts = u_grouped["driver_count"].mean() * 10
u_driverlist = [x for x in u_drivercounts]
s_drivercounts = s_grouped["driver_count"].mean() * 10
s_driverlist = [x for x in s_drivercounts]
r_drivercounts = r_grouped["driver_count"].mean() * 10
r_driverlist = [x for x in r_drivercounts]

# # Build the scatter plots for each city types
u_plot = plt.scatter(u_ridelist, u_farelist, marker="o", facecolors="lightcoral",s=u_driverlist, edgecolors="black", alpha=.8, linewidths=1.5)
s_plot = plt.scatter(s_ridelist, s_farelist, marker="o", facecolors="lightskyblue",s=s_driverlist, edgecolors="black", alpha=0.75, linewidths=1.5)
r_plot = plt.scatter(r_ridelist, r_farelist, marker="o", facecolors="gold",s=r_driverlist, edgecolors="black", alpha=0.75)


# # Incorporate the other graph properties
plt.ylim(19,44)
plt.xlim(1,41)

# Create a legend

# Incorporate a text label regarding circle size
plt.title("Pyber Ride Sharing Data (2016)")
plt.xlabel("Total Number of Rides (Per City)")
plt.ylabel("Average Fare ($)")
string = "Note:\nCircle size correlates with driver count per city"
plt.text(42,30, string)

# Save Figure
plt.grid()
plt.legend((u_plot, s_plot, r_plot), ("Urban", "Suburban", "Rural"))
plt.savefig("Images/BubblePlot.png")
plt.show()
```

### Total Fares by City Type

```python
# Calculate Type Percents
u_farecount = urban["fare"].sum()
s_farecount = suburban["fare"].sum()
r_farecount = rural["fare"].sum()
all_farecount = u_farecount + s_farecount + r_farecount

u_farepct = u_farecount/all_farecount
s_farepct = s_farecount/all_farecount
r_farepct = r_farecount/all_farecount

fare_share = [u_farepct, s_farepct, r_farepct]
# all_count = merged["fare"].count()


# Build Pie Chart
explode = (0.1,0,0)
city_type = ["Urban", "Suburban", "Rural"]
colors = ["lightcoral", "lightskyblue", "gold" ]
plt.pie(fare_share, explode=explode, labels=city_type, colors=colors, autopct="%1.1f%%", startangle=270, shadow=True)
plt.axis("equal")
plt.title("% of Total Fares by City Type")
plt.savefig("Images/Fares By City Type.png")
plt.show
# Save Figure
```

### Total Rides by City Type

```python
# Calculate Ride Percents
u_ridecount = urban["fare"].count()
s_ridecount = suburban["fare"].count()
r_ridecount = rural["fare"].count()
all_ridecount = u_ridecount + s_ridecount + r_ridecount

u_ridepct = u_ridecount/all_ridecount
s_ridepct = s_ridecount/all_ridecount
r_ridepct = r_ridecount/all_ridecount

ride_share = [u_ridepct, s_ridepct, r_ridepct]
# Build Pie Chart
explode = (0.1,0,0)
city_type = ["Urban", "Suburban", "Rural"]
colors = ["lightcoral", "lightskyblue", "gold" ]
plt.pie(ride_share, explode=explode, labels=city_type, colors=colors, autopct="%1.1f%%", startangle=270, shadow=True)
plt.axis("equal")
plt.title("% of Total Rides by City Type")
plt.savefig("Images/Rides By City Type.png")
plt.show

# Save Figure
```

```python
# Calculate Driver Percents
urban_drivers = citydata.loc[citydata["type"] == "Urban"]
suburban_drivers = citydata.loc[citydata["type"] == "Suburban"]
rural_drivers = citydata.loc[citydata["type"] == "Rural"]

u_drivercount = urban_drivers["driver_count"].sum()
s_drivercount = suburban_drivers["driver_count"].sum()
r_drivercount = rural_drivers["driver_count"].sum()
all_drivercount = u_drivercount + s_drivercount + r_drivercount

u_driverpct = u_drivercount/all_drivercount
s_driverpct = s_drivercount/all_drivercount
r_driverpct = r_drivercount/all_drivercount

driver_share = [u_driverpct, s_driverpct, r_driverpct]


# Build Pie Charts
explode = (0.1,0,0)
city_type = ["Urban", "Suburban", "Rural"]
colors = ["lightcoral", "lightskyblue", "gold" ]
plt.pie(driver_share, explode=explode, labels=city_type, colors=colors, autopct="%1.1f%%", startangle=240, shadow=True)
plt.axis("equal")
plt.title("% of Total Drivers by City Type")
plt.savefig("Images/Drivers By City Type.png")
plt.show
# Save Figure
```
